# Tensor Calculus

## Cartesian and Polar Coordinates, and Basis Vectors

- from polar to cartesian
$$\cos(\theta)=\frac{x}{r}\quad x=r\cos(\theta)$$
$$\sin(\theta)=\frac{y}{r}\quad y=r\sin(\theta)$$

- from cartesian to polar
$$r=\sqrt{x^2+y^2}$$
$$\theta=\arctan{(\frac{y}{x})}$$

- Cartesian coordinate basis
$$
\begin{align*}
\frac{\partial{\vec{R}}}{\partial{x}}&=\lim_{h\rightarrow 0}\frac{\overrightarrow{R_h}(x+h,y)-\vec{R}(x,y)}{h}\\
&\equiv\overrightarrow{e_x}
\end{align*}
$$
$$
\begin{align*}
\frac{\partial{\vec{R}}}{\partial{y}}&=\lim_{h\rightarrow 0}\frac{\overrightarrow{R_h}(x,y+h)-\vec{R}(x,y)}{h}\\
&\equiv\overrightarrow{e_y}
\end{align*}
$$

- Polar coordinate basis
$$
\begin{align*}
\frac{\partial{\vec{R}}}{\partial{r}}&=\lim_{h\rightarrow 0}\frac{\overrightarrow{R_h}(r+h,\theta)-\vec{R}(r,\theta)}{h}\\
&\equiv\overrightarrow{e_r}
\end{align*}
$$
$$
\begin{align*}
\frac{\partial{\vec{R}}}{\partial{\theta}}&=\lim_{h\rightarrow 0}\frac{\overrightarrow{R_h}(r,\theta+h)-\vec{R}(r,\theta)}{h}\\
&\equiv\overrightarrow{e_\theta}
\end{align*}
$$

### Forward Transform

$$\widetilde{\overrightarrow{e_r}}=F_r^x\overrightarrow{e_x}+F_r^y\overrightarrow{e_y}$$
$$\widetilde{\overrightarrow{e_\theta}}=F_\theta^x\overrightarrow{e_x}+F_\theta^y\overrightarrow{e_y}$$

$$
\widetilde{\overrightarrow{e_r}}=\frac{\partial \vec{R}}{\partial r}=\frac{\partial x}{\partial r}\frac{\partial \vec{R}}{\partial x}+\frac{\partial y}{\partial r}\frac{\partial \vec{R}}{\partial y}=\frac{\partial x}{\partial r}\overrightarrow{e_x}+\frac{\partial y}{\partial r}\overrightarrow{e_y}
$$
$$
\widetilde{\overrightarrow{e_\theta}}=\frac{\partial \vec{R}}{\partial \theta}=\frac{\partial x}{\partial \theta}\frac{\partial \vec{R}}{\partial x}+\frac{\partial y}{\partial \theta}\frac{\partial \vec{R}}{\partial y}=\frac{\partial x}{\partial \theta}\overrightarrow{e_x}+\frac{\partial y}{\partial \theta}\overrightarrow{e_y}
$$

recall

$$
\begin{align*}
x&=r\cos{\theta}\\
y&=r\sin{\theta}
\end{align*}
\quad\quad\Rightarrow\quad\quad
\begin{align*}
\frac{\partial x}{\partial r}&=\cos{\theta},\quad \frac{\partial x}{\partial\theta}=-r\sin{\theta}\\
\frac{\partial y}{\partial r}&=\sin{\theta},\quad \frac{\partial y}{\partial\theta}=\ r\cos{\theta}
\end{align*}
$$

so

$$
\begin{align*}
\widetilde{\overrightarrow{e_r}}&=\quad\quad\cos{\theta}\ \overrightarrow{e_x}+\quad\quad\sin{\theta}\ \overrightarrow{e_y}\\
\widetilde{\overrightarrow{e_\theta}}&=(-r\sin{\theta})\ \overrightarrow{e_x}+(r\cos{\theta})\ \overrightarrow{e_y}
\end{align*}
$$

$$
F=\begin{bmatrix}
\cos(\theta) & -r\sin(\theta) \\
\sin(\theta) & r\cos(\theta)
\end{bmatrix}=J
$$

$$
F=\begin{bmatrix}
x/r & -y \\
y/r & x
\end{bmatrix}=J
$$

### The Jacobian Matrix (aka "Forward Transform")

$$
F=\begin{bmatrix}
\frac{\partial x}{\partial r} & \frac{\partial x}{\partial \theta} \\
\frac{\partial y}{\partial r} & \frac{\partial y}{\partial \theta}
\end{bmatrix}=J
$$

### Backward Transform

$$
\overrightarrow{e_x}=\frac{\partial \vec{R}}{\partial x}=\frac{\partial r}{\partial x}\frac{\partial \vec{R}}{\partial r}+\frac{\partial \theta}{\partial x}\frac{\partial \vec{R}}{\partial \theta}=\frac{\partial r}{\partial x}\widetilde{\overrightarrow{e_r}}+\frac{\partial \theta}{\partial x}\widetilde{\overrightarrow{e_\theta}}
$$
$$
\overrightarrow{e_y}=\frac{\partial \vec{R}}{\partial y}=\frac{\partial r}{\partial y}\frac{\partial \vec{R}}{\partial r}+\frac{\partial \theta}{\partial y}\frac{\partial \vec{R}}{\partial \theta}=\frac{\partial r}{\partial y}\widetilde{\overrightarrow{e_r}}+\frac{\partial \theta}{\partial y}\widetilde{\overrightarrow{e_\theta}}
$$

$$
\begin{align*}
r&=\sqrt{x^2+y^2}\\
\theta&=\arctan{\frac{y}{x}}
\end{align*}
\quad\quad\Rightarrow\quad\quad
\begin{align*}
\frac{\partial r}{\partial x}&=\frac{x}{\sqrt{x^2+y^2}}=\frac{x}{r},\quad\quad\frac{\partial r}{\partial y}=\frac{y}{\sqrt{x^2+y^2}}=\frac{y}{r}\\
\frac{\partial \theta}{\partial x}&=-\frac{y}{x^2+y^2}=-\frac{y}{r^2},\quad \frac{\partial \theta}{\partial y}=\quad\frac{x}{x^2+y^2}=\frac{x}{r^2}
\end{align*}
$$ 

so

$$
\begin{align*}
\overrightarrow{e_x}&=\frac{x}{r}\ \widetilde{\overrightarrow{e_r}}+(-\frac{y}{r^2})\ \widetilde{\overrightarrow{e_\theta}}\\
\overrightarrow{e_y}&=\frac{y}{r}\ \widetilde{\overrightarrow{e_r}}+\quad\frac{x}{r^2}\quad\widetilde{\overrightarrow{e_\theta}}
\end{align*}
$$

$$
B=\begin{bmatrix}
\frac{x}{r} & \frac{y}{r} \\
-\frac{y}{r^2} & \frac{x}{r^2}
\end{bmatrix}=J^{-1}
$$

### The Inverse Jacobian Matrix (aka "Backward Transform")

$$
B=\begin{bmatrix}
\frac{\partial r}{\partial x} & \frac{\partial r}{\partial y} \\
\frac{\partial\theta}{\partial x} & \frac{\partial\theta}{\partial y}
\end{bmatrix}=J^{-1}
$$

$$
\begin{align*}
J\times B=
\begin{bmatrix}
\frac{\partial x}{\partial r} & \frac{\partial x}{\partial\theta} \\
\frac{\partial y}{\partial r} & \frac{\partial y}{\partial\theta}
\end{bmatrix}
\begin{bmatrix}
\frac{\partial r}{\partial x} & \frac{\partial r}{\partial y} \\
\frac{\partial\theta}{\partial x} & \frac{\partial\theta}{\partial y}
\end{bmatrix}
&=\begin{bmatrix}
\frac{\partial x}{\partial r}\frac{\partial r}{\partial x}+\frac{\partial x}{\partial\theta}\frac{\partial\theta}{\partial x} & \frac{\partial x}{\partial r}\frac{\partial r}{\partial y}+\frac{\partial x}{\partial\theta}\frac{\partial\theta}{\partial y} \\
\frac{\partial y}{\partial r}\frac{\partial r}{\partial x}+\frac{\partial y}{\partial\theta}\frac{\partial\theta}{\partial x} & \frac{\partial y}{\partial r}\frac{\partial r}{\partial y}+\frac{\partial y}{\partial\theta}\frac{\partial\theta}{\partial y}
\end{bmatrix}\\
&=\begin{bmatrix}
\frac{\partial x}{\partial x} & \frac{\partial x}{\partial y} \\ 
\frac{\partial y}{\partial x} & \frac{\partial y}{\partial y} 
\end{bmatrix}=\begin{bmatrix}1 & 0 \\ 0 & 1\end{bmatrix}
\end{align*}
$$
