# Measuring lengths and angles - the first fundamental form

**A curve on a surface.** If a surface $\mathbf{x}$ is parametrized with variables $u$ and $v$, a curve $\gamma$ is on the surface $\mathbf{x}$ if a parametrization of $\gamma$ with a parameter $t$ can be obtained from equations of $\mathbf{x}$ when $u$ and $v$ become functions of $t$
$$
\gamma(t)=\mathbf{x}(u(t),v(t)).
$$

### Examples.

1. A curve given by the equations $\gamma(t)=(x(t),y(t),0)$ is a curve in the plane $z=0$. Recall that this plane can be parametrized by $(x,y,0)$. Thus, in this example, "$u$" is $x$ and "$v$" is $y$.
<br />
<br />
2. The helix $\gamma(t)=(a\cos{t},a\sin{t},bt)$ is on the cylinder $x^2+y^2=a^2$ since the $x$ and $y$ coordinates satisfy the equation of the cylinder. Recall that this cylinder has parametric equations $(a\cos{t},a\sin{t},h)$, so the equations $(a\cos{t},a\sin{t},bt)$ can be perceived as a "special case" of the cylinder equations when one parameter is deleted using the relation $h=bt$. In this example, "$u$" is $t$ and "$v$" is $h$, and, on the helix, $t=t$, and $h=bt$.
<br/>
<br/>
4. **A spherical curve** is a curve that lies on a sphere.

Note that the equation of a sphere $(x−x0)^2+(y−y0)^2+(z−z0)^2=r^2$ has the vector form $((x,y,z)−\mathbf{c})\cdot((x,y,z)−\mathbf{c})=r^2$ where $\mathbf{c}=(x0,y0,z0)$ is the center, $r$ the radius, and $(x,y,z)$ any point on the sphere. Hence, a curve $\gamma(t)=(x(t),y(t),z(t))$ is on this sphere if and only if $(\gamma − \mathbf{c})\cdot(\gamma−\mathbf{c})=r^2$. Differentiating this equation, one obtains the condition that $\gamma'\cdot(\gamma−c)=0$.

Conversely, integrating this last condition (see the proof of Claim 1 in “Curves”), we obtain that $(\gamma−c)\cdot(\gamma−c)$ is constant. Denoting this constant by $r^2$ we obtain an equation of the sphere of radius $r$ centered at $\mathbf{c}$.

For example, $\gamma=(4\cos{2t},4\sin{2t},4\sin{t})$ is spherical for $\mathbf{v}=(−1,0,0)$ since one checks that $\gamma'\cdot(\gamma−c)=0$. The distance from any point on $\gamma$ to $(−1,0,0)$ is the radius of that sphere. So, if we calculate the distance of $(4,0,0)$ with $t=0$, for example, to $(−1,0,1)$, we obtain the radius $r=\sqrt{(4+1)^2+0+0}=5$. Thus, $\gamma$ is on the sphere $(x+1)^2+y^2+z^2=25$.

Surface are often represented by graphing a mesh determined by $u$ and $v$ curves. Because of this, we can think that the mesh of $u$ and $v$ curves forms a **coordinate system** on the surface.

### Examples.
1. On the sphere $\mathbf{x}=(a\cos{\theta}\cos{\phi},a\sin{\theta}\cos{\phi},a\sin{\phi}$, the $\phi$-curves are circles of constant longitude **meridians** and the $\theta$-curves are circles of constant latitude, **parallels**.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline

theta, phi = np.meshgrid(np.linspace(0, 2*np.pi, 60), np.linspace(-np.pi/2, np.pi/2, 30))
a = 1.0

x, y, z = a*np.cos(theta)*np.cos(phi), a*np.sin(theta)*np.cos(phi), a*np.sin(phi)

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(projection='3d')

ax.plot_surface(x, y, z, cmap=cm.viridis, alpha=0.8)
ax.plot_wireframe(x, y, z, linewidth=0.5, rcount=30, ccount=60)
ax.set_box_aspect((np.ptp(x), np.ptp(y), np.ptp(z)))

plt.show()
```

2. **Helicoid**. The surface parametrized by $(r\cos{\theta},r\sin{\theta},a\theta)$ has $\theta$-curves helices spiraling about cylinder of radius $r_0$ and $r$-curves lines $(r\cos{\theta_0},r\sin{\theta_0},a\theta_0)$ passing through $z$-axis in a plane parallel to $xy$-plane. The surface resembles the spiral ramps like those found in garages.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline

theta, r = np.meshgrid(np.linspace(-2*np.pi, 3.5*np.pi, 100), np.linspace(0, 1, 30))
a = 5.0

x, y, z = r * np.cos(theta), r * np.sin(theta), a * theta

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(projection='3d')

ax.plot_surface(x, y, z, cmap=cm.viridis, alpha=1)
ax.plot_wireframe(x, y, z, linewidth=0.5, rcount=30, ccount=100)
# ax.set_box_aspect((np.ptp(x), np.ptp(y), np.ptp(z)))

plt.show()
```

3. **Surfaces of revolution**. A surface of revolution of curve $\alpha=(r(t),z(t))$ about $z$-axis can be given by $(r(t)\cos{\theta},r(t)\sin{\theta},z(t))$. The $\theta$-curves are circles of radii $r(t0)$ in horizontal planes passing $z(t0)$. They are also called the **circles of latitude** or the **parallels** by analogy with parallels on a sphere. The $t$-curves have the same shape as the curve $\alpha$ except that they are positioned in vertical planes at longitude $\theta_0$. They are called the meridians.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

t = np.linspace(1, 5, 50)
r = lambda t: t
zf = lambda t: 5 / t

t, theta = np.meshgrid(t, np.linspace(0, 2*np.pi, 50))

x, y, z = r(t)*np.cos(theta), r(t)*np.sin(theta), zf(t)

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(projection='3d')

ax.plot_surface(x, y, z, cmap=cm.viridis, alpha=0.8)
ax.set_box_aspect((np.ptp(x), np.ptp(y), np.ptp(z)))
plt.show()
```

### Examples of ruled surfaces.
1. A plane can be considered to be a ruled surface letting $\mathbf{\alpha}(t)$ be a line and $\mathbf{\beta}(t)$ be a constant vector.
<br />
<br />
2. A **cone** is a ruled surface with $\mathbf{\alpha}$ constant, say point $P$. The point $P$ is called the vertex of the cone. In case that $\mathbf{\beta}$ makes a constant angle with fixed line through $P$ (called the **axis**) of the cone), we obtain the right circular cone. For example, the cone $z=\sqrt{x^2+y^2}$ can be parametrized by taking $\mathbf{\alpha}=(0,0,0)$ and $\mathbf{\beta}=(\cos{t},\sin{t},1)$ and getting $(s\cos{t},s\sin{t}, s)$.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

t, s = np.meshgrid(np.linspace(0, 2*np.pi, 100), np.linspace(0, 3, 100))

x, y, z = s*np.cos(t), s*np.sin(t), s

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(projection='3d')

ax.plot_surface(x, y, z, cmap=cm.viridis, alpha=0.8)
ax.set_box_aspect((np.ptp(x), np.ptp(y), np.ptp(z)))
plt.show()
```

3. A **cylindrical surface** is defined as a ruled surface with $\mathbf{\beta}$ constant vector. If $\mathbf{\alpha}$ is a circle, the cylindrical surface is said to be **circular cylinder**. If $\mathbf{\beta}$ is a vector perpendicular to the plane of circle $\mathbf{\alpha}$ the circular cylinder is said to be **right**.

For example, the cylinder $x^2+y^2=a^2$ can be considered as a ruled surfaces with $\mathbf{\alpha}$ being the circle in $xy$-plane and $\mathbf{\beta}=(0,0,1)$.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

t, s = np.meshgrid(np.linspace(0, 2*np.pi, 100), np.linspace(0, 2, 100))
r = 1.0

alpha = np.array([r * np.cos(t), r*np.sin(t), np.zeros_like(t)])
beta = np.array([0, 0, 1])

X = np.zeros_like(alpha)
for i in range(beta.shape[0]):
    X[i] = alpha[i] + s * beta[i]

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(projection='3d')

ax.plot_surface(X[0], X[1], X[2], cmap=cm.viridis, alpha=0.8)
ax.set_box_aspect((np.ptp(X[0]), np.ptp(X[1]), np.ptp(X[2])))
plt.show()
```

## The First Fundamental Form

**Example 1. The first fundamental form of a plane.** Consider the $xy$-plane $z=0$ for simplicity. Thus $\mathbf{x}(x,y)=(x,y,0)$ and you can consider that "$u$" is $x$ and "$v$" is $y$ here.

```python
from sympy import *

x, y = symbols('x, y')
init_printing()

gamma = Matrix([x, y, 0])
# the basis vectors of the tangent plane
gamma_1 = diff(gamma, x)
gamma_2 = diff(gamma, y)
print('gamma_1 = ', gamma_1)
print('gamma_2 = ', gamma_2)

# the coefficients of the first fundamental form
g11 = gamma_1.dot(gamma_1)
g12 = gamma_1.dot(gamma_2)
g22 = gamma_2.dot(gamma_2)
print('g11 = ', g11)
print('g12 = ', g12)
print('g22 = ', g22)

t, a, b = symbols('t, a, b')
x = Function('x')(t)
y = Function('y')(t)

# the arc-length
L = simplify(integrate(sqrt(g11*diff(x, t)**2 + 2*g12*diff(x, t)
             * diff(y, t) + g22*diff(y, t)**2), (t, a, b)))
L
```

**Example 2. The first fundamental form of a cylinder.** Consider the cylinder $(a\cos{t},a\sin{t},z)$ with a circular base of radius $a$. Thus you can consider that "$u$" is $t$ and "$v$" is $z$.

```python
from sympy import *

a, b = symbols('a, b', positive=True)
t, z = symbols('t, z')
init_printing()

gamma = Matrix([a*cos(t), a*sin(t), z])
gamma_1 = diff(gamma, t)
gamma_2 = diff(gamma, z)
print('gamma_1 = ', gamma_1)
print('gamma_2 = ', gamma_2)

g11 = simplify(gamma_1.dot(gamma_1))
g12 = gamma_1.dot(gamma_2)
g22 = gamma_2.dot(gamma_2)
print('g11 = ', g11)
print('g12 = ', g12)
print('g22 = ', g22)
```

**Example 3. The first fundamental form of a sphere.** Consider the sphere of radius $a$ centered at the origin. Using the geographical coordinates
$$
\mathbf{x}=(a\cos{\theta}\cos{\phi},a\sin{\theta}\cos{\phi},a\sin{\phi}).
$$
If you consider that "$u$" is $\theta$ and "$v$" is $\phi$:

```python
from sympy import *

a = symbols('a', positive=True)
theta, phi = symbols('theta, phi')
init_printing()

gamma = Matrix([a*cos(theta)*cos(phi), a*sin(theta)*cos(phi), a*sin(phi)])

gamma_1 = diff(gamma, theta)
gamma_2 = diff(gamma, phi)
print('gamma_1 = ', gamma_1)
print('gamma_2 = ', gamma_2)

g11 = simplify(gamma_1.dot(gamma_1))
g12 = gamma_1.dot(gamma_2)
g22 = simplify(gamma_2.dot(gamma_2))
print('g11 = ', g11)
print('g12 = ', g12)
print('g22 = ', g22)
```

**Example 4. The first fundamental form of the surface** $z=z(x,y)$. Let us consider this surface as $\mathbf{x}=(x,y,z(x,y))$ parametrized by $x$ and $y$.

```python
from sympy import *

x, y = symbols('x, y')
z = Function('z')(x, y)
init_printing()

gamma = Matrix([x, y, z])
gamma_1 = diff(gamma, x)
gamma_2 = diff(gamma, y)
print('gamma_1 = ', gamma_1)
print('gamma_2 = ', gamma_2)

g11 = simplify(gamma_1.dot(gamma_1))
g12 = gamma_1.dot(gamma_2)
g22 = simplify(gamma_2.dot(gamma_2))
print('g11 = ', g11)
print('g12 = ', g12)
print('g22 = ', g22)
```

2. Find the surface area of the part of the cylinder $\mathbf{x}=(a\cos{\theta},a\sin{\theta},h)$ with $0\le z\le c$.

**Solution**

In one of the previous examples, we computed the coefficients of the first fundamental form of the cylinder $\mathbf{x}=(a\cos{\theta},a\sin{\theta},h)$ to be $g_{11}=a^2,g_{12}=0,g_{22}=1$. Hence $g=g_{11}g_{22}-g_{12}^2=a^2$.

Since $\sqrt{g}=a$, the formula $S=\int\int_S{a\ d\theta\ dh}$ computes the surface area. On the region with $0\le z\le c$, we have that $0\le\theta\le2\pi$ and $0\le h\le c$. Hence,
$$S=\int_0^{2\pi}\int_0^c{a\ d\theta\ dh}=a\int_0^{2\pi}d\theta\int_0^c dh=a(2\pi)c=2ac\pi$$

3. Consider the torus obtained by revolving a circle $(x-a)^2+z^2=b^2$ in $xz$-plane along the circle $x^2+y^2=a^2$ in $xy$-plane. Since the first circle can be parametrized by $x=a+b\cos{\phi},z=b\sin{\phi}$ and a surface of revolution of a curve $x=f(u),z=g(u)$ in $xz$-plane about $z$-axis is given by the parametric equation $\mathbf{x}=(f(u)\cos{\theta},f(u)\sin{\theta},g(u))$, the torus can be parametrized as

4. Find the area of the part of the parboloid $z=a^2-x^2-y^2,a>0$, that lies above the $xy$-plane.

**Solution**

Considering the paraboloid as a surface of revolution produces the parametrization
$$
\mathbf{x}=(r\cos{t},r\sin{t},a^2-r^2)
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

t, r = np.meshgrid(np.linspace(0, 2*np.pi, 100), np.linspace(0, 1, 20))

a = 1.0
x, y, z = (r*np.cos(t), r*np.sin(t), a**2-r**2)

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(projection='3d')

ax.plot_surface(x, y, z, cmap=cm.viridis, alpha=0.8)
ax.set_box_aspect((np.ptp(x), np.ptp(y), np.ptp(z)))
plt.show()
```

```python
from sympy import *

a, r = symbols('a, r', positive=True, real=True)
t = symbols('t', real=True)
init_printing()

gamma = Matrix([r*cos(t), r*sin(t), a**2-r**2])
gamma_1 = diff(gamma, r)
gamma_2 = diff(gamma, t)
print('gamma_1 = ', gamma_1)
print('gamma_2 = ', gamma_2)

g11 = simplify(gamma_1.dot(gamma_1))
g12 = gamma_1.dot(gamma_2)
g22 = simplify(gamma_2.dot(gamma_2))
print('g11 = ', g11)
print('g12 = ', g12)
print('g22 = ', g22)

g = g11*g22 - g12**2
print('g = ', g)
```

The paraboloid intersects the $xy$-plane $z=0$ at the circle with radius $a^2-r^2=0\implies r^2=a^2\implies r=a$. Hence the bounds are $0\le r\le a$ and $0\le t\le 2\pi$. The area is
$$
S=\int_0^{2\pi}\int_0^a r\sqrt{1+4r^2}dr\ dt=\left.2\pi\frac{1}{12}(1+4r^2)^{3/2}\right\vert_0^a=\frac{\pi}{6}((1+4a^2)^{3/2}-1)
$$

```python
S = integrate(integrate(sqrt(g), (r, 0, a)), (t, 0, 2*pi))
print(simplify(pi/6*(sqrt(1+4*a**2)**3-1) - S) == 0)
S
```

5. Find the area of the part of the cone $z=a\sqrt{x^2+y^2}$ below the plane $z=b$ where $a,b>0$.

**Solution**

Considering the cone as a surface of revolution produces the parametrization
$$
\mathbf{x}=(r\cos{t}, r\sin{t}, ar)
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline

t, r = np.meshgrid(np.linspace(0, 2*np.pi, 100), np.linspace(0, 1, 20))

a = 1.0
x, y, z = (r*np.cos(t), r*np.sin(t), a*r)

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(projection='3d')

ax.plot_surface(x, y, z, cmap=cm.viridis, alpha=0.8)
ax.set_box_aspect((np.ptp(x), np.ptp(y), np.ptp(z)))

b = 0.7
x, y = np.meshgrid(np.linspace(-1, 1, 50), np.linspace(-1, 1, 50))
z = np.ones_like(x) * b

ax.plot_wireframe(x, y, z, linewidth=0.6, alpha=1)

plt.show()
```

```python
from sympy import *

a, r = symbols('a, r', positive=True, real=True)
t = symbols('t', real=True)
init_printing()

gamma = Matrix([r*cos(t), r*sin(t), a*r])
gamma_1 = diff(gamma, r)
gamma_2 = diff(gamma, t)
print('gamma_1 = ', gamma_1)
print('gamma_2 = ', gamma_2)

g11 = simplify(gamma_1.dot(gamma_1))
g12 = gamma_1.dot(gamma_2)
g22 = simplify(gamma_2.dot(gamma_2))
print('g11 = ', g11)
print('g12 = ', g12)
print('g22 = ', g22)

g = g11*g22 - g12**2
print('g = ', g)
```

The cone intersects the $xy$-plane $z=b$ at the circle with radius $ar=b\implies r=\frac{b}{a}$. Hence the bounds are $0\le r\le \frac{b}{a}$ and $0\le t\le 2\pi$. The area is
$$
S=\int_0^{2\pi}\int_0^{b/a}r\sqrt{1+a^2}dr\ dt=\sqrt{1+a^2}\int_0^{2\pi}dt\int_0^{b/a}r\ dr=\sqrt{1+a^2}2\pi\frac{b^2}{2a^2}=\frac{b^2\pi\sqrt{1+a^2}}{a^2}
$$

```python
b = symbols('b', positive=True, real=True)

S = integrate(integrate(sqrt(g), (r, 0, b/a)), (t, 0, 2*pi))
print(simplify(b**2*pi*sqrt(1+a**2)/a**2) - S == 0)
S
```

(b) With $\mathbf{\beta}=(a\cos{t},a\sin{t},a^2)$,
$$
\mathbf{x}=\mathbf{\alpha}+s\mathbf{\beta}=(0,0,0)+(as\cos{t},as\sin{t},a^2s)=(as\cos{t},as\sin{t},a^2s)
$$

```python
from sympy import *

a, s = symbols('a, s', positive=True, real=True)
t = symbols('t', real=True)
init_printing()

gamma = Matrix([a*s*cos(t), a*s*sin(t), a**2*s])
gamma_1 = diff(gamma, s)
gamma_2 = diff(gamma, t)
print('gamma_1 = ', gamma_1)
print('gamma_2 = ', gamma_2)

g11 = simplify(gamma_1.dot(gamma_1))
g12 = gamma_1.dot(gamma_2)
g22 = simplify(gamma_2.dot(gamma_2))
print('g11 = ', g11)
print('g12 = ', g12)
print('g22 = ', g22)

g = g11*g22 - g12**2
print('g = ', g)
```

The cone intersects the $xy$-plane $z=b$ at the circle with radius $a^2s=b\implies s=\frac{b}{a^2}$. Hence the bounds are $0\le s\le \frac{b}{a^2}$ and $0\le t\le 2\pi$. The area is
$$
S=\int_0^{2\pi}\int_0^{b/a^2}a^2s\sqrt{1+a^2}ds\ dt=a^2\sqrt{1+a^2}\int_0^{2\pi}dt\int_0^{b/a^2}s\ dr=a^2\sqrt{1+a^2}2\pi\frac{b^2}{2a^4}=\frac{b^2\pi\sqrt{1+a^2}}{a^2}
$$
which is the same as with the other parametrization. The previous two problems illustrate that the surface area is independent of specific parametrization.

```python
b = symbols('b', positive=True, real=True)

S = integrate(integrate(sqrt(g), (s, 0, b/a**2)), (t, 0, 2*pi))
print(simplify(b**2*pi*sqrt(1+a**2)/a**2) - S == 0)
S
```
